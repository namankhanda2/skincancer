```python
# Cryptocurrency Price Prediction using Time Series Analysis
# ARIMA Model Implementation

# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
from datetime import datetime, timedelta
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.stats.diagnostic import acorr_ljungbox
from statsmodels.stats.stattools import jarque_bera
from sklearn.metrics import mean_squared_error, mean_absolute_error
import statsmodels.api as sm
import warnings
import math
import io
import matplotlib.dates as mdates

# Set plotting style and ignore warnings
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (14, 8)
plt.rcParams['font.size'] = 14
warnings.filterwarnings('ignore')

# Function to calculate Mean Absolute Percentage Error (MAPE)
def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    # Handle zero values to avoid division by zero
    mask = y_true > 0
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100

# Function to calculate directional accuracy
def directional_accuracy(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    # Calculate direction of movements
    y_true_dir = np.sign(np.diff(y_true))
    y_pred_dir = np.sign(np.diff(y_pred))
    # Calculate accuracy
    return np.mean(y_true_dir == y_pred_dir) * 100

# Function to calculate Theil's U statistic
def theil_u(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    # Calculate squared error
    squared_error = np.sum((y_pred - y_true) ** 2)
    # Calculate naive forecast squared error (using previous value as forecast)
    naive_error = np.sum((y_true[1:] - y_true[:-1]) ** 2)
    return np.sqrt(squared_error / naive_error)

# Function to perform ADF test and print results
def check_stationarity(timeseries, title=''):
    print(f"Augmented Dickey-Fuller Test for {title}")
    result = adfuller(timeseries.dropna())

    print(f'ADF Statistic: {result[0]:.4f}')
    print(f'p-value: {result[1]:.4f}')
    print('Critical Values:')
    for key, value in result[4].items():
        print(f'   {key}: {value:.4f}')

    # Interpret the result
    if result[1] <= 0.05:
        print("Result: The series is stationary (reject H0)")
    else:
        print("Result: The series is non-stationary (fail to reject H0)")
    print("\n")

    return result[1] <= 0.05  # Return True if stationary

# Function to plot time series data
def plot_time_series(series, title='', xlabel='Date', ylabel='Price'):
    plt.figure(figsize=(14, 8))
    plt.plot(series.index, series.values)
    plt.title(title, fontsize=16)
    plt.xlabel(xlabel, fontsize=14)
    plt.ylabel(ylabel, fontsize=14)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Function to find optimal ARIMA parameters using AIC
def find_optimal_arima(timeseries, p_range, d, q_range):
    best_aic = float("inf")
    best_params = None
    best_model = None

    # Create a table to store results
    results_table = []

    for p in p_range:
        for q in q_range:
            try:
                model = ARIMA(timeseries, order=(p, d, q))
                model_fit = model.fit()
                aic = model_fit.aic
                bic = model_fit.bic
                results_table.append((p, d, q, aic, bic))

                if aic < best_aic:
                    best_aic = aic
                    best_params = (p, d, q)
                    best_model = model_fit

            except Exception as e:
                continue

    # Convert results to DataFrame and sort by AIC
    results_df = pd.DataFrame(results_table, columns=['p', 'd', 'q', 'AIC', 'BIC'])
    results_df = results_df.sort_values('AIC')

    print("Top 5 ARIMA Models by AIC:")
    print(results_df.head(5).to_string(index=False))
    print(f"\nOptimal ARIMA Parameters: {best_params} with AIC: {best_aic:.2f}")

    return best_params, best_model, results_df

# Function to plot model diagnostics
def plot_diagnostics(model_fit):
    plt.figure(figsize=(16, 12))

    # Plot residuals
    plt.subplot(321)
    plt.plot(model_fit.resid)
    plt.title('Residuals')
    plt.grid(True)

    # Plot histogram of residuals
    plt.subplot(322)
    plt.hist(model_fit.resid, bins=30)
    plt.title('Histogram of Residuals')

    # Plot ACF of residuals
    plt.subplot(323)
    plot_acf(model_fit.resid, lags=40, ax=plt.gca())
    plt.title('ACF of Residuals')

    # Plot PACF of residuals
    plt.subplot(324)
    plot_pacf(model_fit.resid, lags=40, ax=plt.gca())
    plt.title('PACF of Residuals')

    # QQ plot
    plt.subplot(325)
    sm.qqplot(model_fit.resid, line='45', fit=True, ax=plt.gca())
    plt.title('QQ Plot')

    # Plot residuals vs fitted values
    plt.subplot(326)
    plt.scatter(model_fit.fittedvalues, model_fit.resid)
    plt.axhline(y=0, color='red', linestyle='--')
    plt.title('Residuals vs Fitted Values')
    plt.xlabel('Fitted Values')
    plt.ylabel('Residuals')

    plt.tight_layout()
    plt.show()

# Function to evaluate forecast performance
def evaluate_forecast(y_true, y_pred):
    rmse = math.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    mape = mean_absolute_percentage_error(y_true, y_pred)
    dir_acc = directional_accuracy(y_true, y_pred)
    theil = theil_u(y_true, y_pred)

    print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
    print(f"Mean Absolute Error (MAE): {mae:.2f}")
    print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")
    print(f"Directional Accuracy: {dir_acc:.2f}%")
    print(f"Theil's U Statistic: {theil:.2f}")

    return {
        'RMSE': rmse,
        'MAE': mae,
        'MAPE': mape,
        'Directional Accuracy': dir_acc,
        'Theil U': theil
    }

# Function to plot actual vs forecasted values
def plot_forecast(actual, forecast, conf_int=None, title='Actual vs Forecast'):
    plt.figure(figsize=(14, 8))

    # Plot actual values
    plt.plot(actual.index, actual, 'b-', label='Actual')

    # Plot forecasted values
    plt.plot(forecast.index, forecast, 'r-', label='Forecast')

    # Plot confidence intervals if provided
    if conf_int is not None:
        plt.fill_between(forecast.index,
                         conf_int.iloc[:, 0],
                         conf_int.iloc[:, 1],
                         color='pink', alpha=0.3, label='95% Confidence Interval')

    plt.title(title, fontsize=16)
    plt.xlabel('Date', fontsize=14)
    plt.ylabel('Price', fontsize=14)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Function to plot forecast errors
def plot_forecast_errors(actual, forecast, title='Forecast Errors'):
    errors = actual - forecast

    plt.figure(figsize=(14, 6))
    plt.plot(errors.index, errors, 'r-')
    plt.axhline(y=0, color='black', linestyle='--')
    plt.title(title, fontsize=16)
    plt.xlabel('Date', fontsize=14)
    plt.ylabel('Error', fontsize=14)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Function to implement and evaluate benchmark models
def evaluate_benchmark_models(train_data, test_data):
    benchmarks = {}

    # 1. Naive (Random Walk) Model
    naive_forecast = train_data.iloc[-1]
    naive_predictions = pd.Series([naive_forecast] * len(test_data), index=test_data.index)
    naive_metrics = evaluate_forecast(test_data, naive_predictions)
    benchmarks['Naive'] = naive_metrics

    # 2. Simple Moving Average (5-day)
    sma_window = 5
    sma_values = train_data.rolling(window=sma_window).mean().iloc[-1]
    sma_predictions = pd.Series([sma_values] * len(test_data), index=test_data.index)
    sma_metrics = evaluate_forecast(test_data, sma_predictions)
    benchmarks['SMA'] = sma_metrics

    # 3. Exponential Smoothing
    alpha = 0.3  # Smoothing factor
    es_model = sm.tsa.ExponentialSmoothing(train_data).fit()
    es_predictions = es_model.forecast(len(test_data))
    es_metrics = evaluate_forecast(test_data, es_predictions)
    benchmarks['ES'] = es_metrics

    # Create comparison table
    benchmark_df = pd.DataFrame(benchmarks).T

    print("\nBenchmark Models Comparison:")
    print(benchmark_df)

    return benchmark_df

# Main execution
if __name__ == "__main__":
    print("Cryptocurrency Price Prediction using Time Series Analysis")
    print("========================================================")

    # Step 1: Data Collection
    print("\n1. Collecting Data...")

    # Define the date range
    start_date = '2018-01-01'
    end_date = '2025-03-31'  # Adjust as needed

    # Download Bitcoin historical data
    ticker = 'BTC-USD'
    try:
        # Set auto_adjust to False to get both Close and Adj Close
        df = yf.download(ticker, start=start_date, end=end_date, group_by='Ticker')
        print(f"Successfully downloaded {len(df)} days of {ticker} data from {df.index.min().date()} to {df.index.max().date()}")
    except Exception as e:
        print(f"Error downloading data: {e}")
        # If download fails, try with a more recent end date
        current_date = datetime.now()
        end_date = current_date.strftime('%Y-%m-%d')
        print(f"Retrying with end date: {end_date}")
        df = yf.download(ticker, start=start_date, end=end_date, group_by='Ticker')
        print(f"Successfully downloaded {len(df)} days of {ticker} data from {df.index.min().date()} to {df.index.max().date()}")

    # Display the first few rows of the dataset
    print("\nData Preview:")
    print(df.head())

    # Check data info
    print("\nData Information:")
    print(f"Data shape: {df.shape}")
    print("\nColumn Data Types:")
    print(df.dtypes)
    print("\nMissing Values:")
    print(df.isnull().sum())

    # Step 2: Data Preprocessing
    print("\n2. Preprocessing Data...")

    # Handle missing values
    if df.isnull().sum().sum() > 0:
        print("Handling missing values using forward fill method...")
        df = df.fillna(method='ffill')
        print("Missing values after handling:")
        print(df.isnull().sum())

    # Print column names to check structure
    print("\nAvailable columns in the dataframe:")
    print(df.columns.tolist())

    # Check if we have a MultiIndex DataFrame
    is_multiindex = isinstance(df.columns, pd.MultiIndex)
    print(f"DataFrame has MultiIndex columns: {is_multiindex}")

    # Handle the price data extraction based on the DataFrame structure
    if is_multiindex:
        # For MultiIndex, we need to select the Close price for the ticker
        print("Extracting price data from MultiIndex DataFrame...")
        if ('Close', ticker) in df.columns:
            prices = df[('Close', ticker)].copy()
            print(f"Using 'Close' price for {ticker}")
        elif ('Adj Close', ticker) in df.columns:
            prices = df[('Adj Close', ticker)].copy()
            print(f"Using 'Adj Close' price for {ticker}")
        else:
            # If neither Close nor Adj Close exists with the ticker, try the first level only
            available_cols = [col[0] for col in df.columns]
            if 'Close' in available_cols:
                prices = df['Close'].copy()
                print("Using 'Close' price column")
            else:
                # Last resort: just take the first price column
                first_col = df.columns[0]
                prices = df[first_col].copy()
                print(f"Using {first_col} price column")
    else:
        # For standard DataFrame, just check if Close or Adj Close exists
        if 'Close' in df.columns:
            prices = df['Close'].copy()
            print("Using 'Close' price column")
        elif 'Adj Close' in df.columns:
            prices = df['Adj Close'].copy()
            print("Using 'Adj Close' price column")
        else:
            # If neither exists, use the first column
            first_col = df.columns[0]
            prices = df[first_col].copy()
            print(f"Using {first_col} price column")

    # Ensure prices is a Series, not a DataFrame
    if isinstance(prices, pd.DataFrame):
        print("Converting prices from DataFrame to Series...")
        prices = prices.iloc[:, 0]

    # Verify the type of prices
    print(f"\nPrice data type: {type(prices)}")
    print(f"Number of price observations: {len(prices)}")

    # Calculate descriptive statistics
    print("\nDescriptive Statistics for Bitcoin Prices:")
    stats = prices.describe()
    print(stats)

    # Additional statistics
    skewness = prices.skew()
    kurtosis = prices.kurtosis()

    # Use pandas Series autocorr method directly, which should exist for Series
    try:
        first_order_autocorr = prices.autocorr(lag=1)
        print(f"\nSkewness: {skewness:.2f}")
        print(f"Kurtosis: {kurtosis:.2f}")
        print(f"First-Order Autocorrelation: {first_order_autocorr:.4f}")
    except AttributeError:
        print("\nCould not calculate autocorrelation. Continuing with analysis.")
        print(f"Skewness: {skewness:.2f}")
        print(f"Kurtosis: {kurtosis:.2f}")

    # Plot the time series data
    plot_time_series(prices, title=f'{ticker} Price History', ylabel='Price (USD)')

    # Check for stationarity
    print("\nChecking stationarity of the original series:")
    original_stationary = check_stationarity(prices, title='Original Series')

    # Apply differencing if not stationary
    if not original_stationary:
        prices_diff = prices.diff().dropna()
        print("Checking stationarity after first-order differencing:")
        diff_stationary = check_stationarity(prices_diff, title='Differenced Series')

        # Plot differenced series
        plot_time_series(prices_diff, title=f'{ticker} Price (First Difference)', ylabel='Price Change (USD)')

        # Plot ACF and PACF of differenced series
        plt.figure(figsize=(14, 8))
        plt.subplot(211)
        plot_acf(prices_diff, lags=40, ax=plt.gca())
        plt.title('Autocorrelation Function (ACF)')

        plt.subplot(212)
        plot_pacf(prices_diff, lags=40, ax=plt.gca())
        plt.title('Partial Autocorrelation Function (PACF)')
        plt.tight_layout()
        plt.show()

        # Set differencing order based on stationarity test
        d_order = 1
    else:
        d_order = 0

    # Step 3: Split data into training and testing sets
    print("\n3. Splitting data into training and testing sets...")

    # Use 90% for training and 10% for testing
    train_size = int(len(prices) * 0.9)
    train_data = prices[:train_size]
    test_data = prices[train_size:]

    print(f"Training data: {train_data.index.min().date()} to {train_data.index.max().date()} ({len(train_data)} observations)")
    print(f"Testing data: {test_data.index.min().date()} to {test_data.index.max().date()} ({len(test_data)} observations)")

    # Plot the train-test split
    plt.figure(figsize=(14, 8))
    plt.plot(train_data.index, train_data, 'b-', label='Training Data')
    plt.plot(test_data.index, test_data, 'r-', label='Testing Data')
    plt.axvline(x=train_data.index[-1], color='black', linestyle='--')
    plt.title(f'{ticker} Train-Test Split', fontsize=16)
    plt.xlabel('Date', fontsize=14)
    plt.ylabel('Price (USD)', fontsize=14)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Step 4: Find optimal ARIMA parameters
    print("\n4. Finding optimal ARIMA parameters...")

    # Set range for p and q
    p_range = range(0, 4)
    q_range = range(0, 4)

    # Find optimal parameters
    best_params, best_model, results_df = find_optimal_arima(train_data, p_range, d_order, q_range)

    # Display model summary
    print("\nARIMA Model Summary:")
    print(best_model.summary())

    # Step 5: Model Diagnostics
    print("\n5. Performing model diagnostics...")

    # Plot diagnostic plots
    plot_diagnostics(best_model)

    # Ljung-Box test for autocorrelation in residuals
    lb_test = acorr_ljungbox(best_model.resid, lags=[10, 20, 30], return_df=True)
    print("\nLjung-Box Test for Residual Autocorrelation:")
    print(lb_test)

    # Jarque-Bera test for normality of residuals
    jb_test = jarque_bera(best_model.resid)
    print("\nJarque-Bera Test for Residual Normality:")
    print(f"Statistic: {jb_test[0]:.4f}")
    print(f"p-value: {jb_test[1]:.4f}")
    if jb_test[1] > 0.05:
        print("Result: Residuals are normally distributed")
    else:
        print("Result: Residuals are not normally distributed")

    # Step 6: In-sample forecasting and evaluation
    print("\n6. In-sample forecasting and evaluation...")

    # Get fitted values
    in_sample_predictions = best_model.fittedvalues

    # Plot actual vs fitted values
    plt.figure(figsize=(14, 8))
    plt.plot(train_data.index, train_data, 'b-', label='Actual')
    plt.plot(in_sample_predictions.index, in_sample_predictions, 'r-', label='Fitted')
    plt.title(f'In-Sample Fit: ARIMA{best_params}', fontsize=16)
    plt.xlabel('Date', fontsize=14)
    plt.ylabel('Price (USD)', fontsize=14)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Evaluate in-sample performance
    print("\nIn-Sample Performance Metrics:")
    in_sample_metrics = evaluate_forecast(train_data, in_sample_predictions)

    # Step 7: Out-of-sample forecasting and evaluation
    print("\n7. Out-of-sample forecasting...")

    # Generate forecasts
    forecast_steps = len(test_data)
    forecast_result = best_model.get_forecast(steps=forecast_steps)
    forecast_mean = forecast_result.predicted_mean
    forecast_conf_int = forecast_result.conf_int(alpha=0.05)  # 95% confidence interval

    # Set proper index for forecasts
    forecast_mean.index = test_data.index
    forecast_conf_int.index = test_data.index

    # Plot forecasts with confidence intervals
    plot_forecast(test_data, forecast_mean, forecast_conf_int,
                 title=f'Bitcoin Price Forecast: ARIMA{best_params}')

    # Plot forecast errors
    plot_forecast_errors(test_data, forecast_mean, title='Forecast Errors')

    # Evaluate out-of-sample performance
    print("\nOut-of-Sample Performance Metrics:")
    out_sample_metrics = evaluate_forecast(test_data, forecast_mean)

    # Step 8: Compare with benchmark models
    print("\n8. Comparing with benchmark models...")
    benchmark_results = evaluate_benchmark_models(train_data, test_data)

    # Add ARIMA results to the comparison
    all_results = benchmark_results.copy()
    all_results.loc['ARIMA'] = pd.Series(out_sample_metrics)

    # Plot comparison of metrics
    metrics_to_plot = ['RMSE', 'MAE', 'MAPE']
    plt.figure(figsize=(14, 8))

    for i, metric in enumerate(metrics_to_plot):
        plt.subplot(1, 3, i+1)
        ax = all_results[metric].plot(kind='bar', color='skyblue')
        plt.title(metric)
        plt.grid(True, axis='y')

        # Add value labels on bars
        for p in ax.patches:
            ax.annotate(f"{p.get_height():.2f}",
                        (p.get_x() + p.get_width() / 2., p.get_height()),
                        ha = 'center', va = 'bottom')

    plt.tight_layout()
    plt.show()

    # Step 9: Generate a full forecast
    print("\n9. Generating full forecast for the next 30 days...")

    # Create a final model using all data
    final_model = ARIMA(prices, order=best_params).fit()

    # Generate forecast for next 30 days
    future_steps = 30
    future_forecast = final_model.get_forecast(steps=future_steps)
    future_mean = future_forecast.predicted_mean
    future_conf_int = future_forecast.conf_int(alpha=0.05)

    # Create date range for future forecasts
    last_date = prices.index[-1]
    future_dates = pd.date_range(start=last_date + timedelta(days=1), periods=future_steps)

    # Set proper index
    future_mean.index = future_dates
    future_conf_int.index = future_dates

    # Plot final forecast
    plt.figure(figsize=(14, 8))

    # Plot historical data (last 180 days)
    plt.plot(prices[-180:].index, prices[-180:], 'b-', label='Historical Prices')

    # Plot forecast
    plt.plot(future_mean.index, future_mean, 'r-', label='Forecast')

    # Plot confidence intervals
    plt.fill_between(future_mean.index,
                     future_conf_int.iloc[:, 0],
                     future_conf_int.iloc[:, 1],
                     color='pink', alpha=0.3, label='95% Confidence Interval')

    plt.title(f'Bitcoin 30-Day Price Forecast: ARIMA{best_params}', fontsize=16)
    plt.xlabel('Date', fontsize=14)
    plt.ylabel('Price (USD)', fontsize=14)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Display forecast values
    print("\nForecasted Bitcoin Prices for the Next 30 Days:")
    forecast_df = pd.DataFrame({
        'Date': future_dates,
        'Forecast': future_mean.values,
        'Lower CI': future_conf_int.iloc[:, 0].values,
        'Upper CI': future_conf_int.iloc[:, 1].values
    })
    forecast_df['Date'] = forecast_df['Date'].dt.date
    forecast_df = forecast_df.set_index('Date')
    print(forecast_df)

    # Step 10: Save the results
    print("\n10. Saving results...")

    # Create a buffer to capture plots for the report
    buffer = io.BytesIO()

    # Save forecast data to CSV
    forecast_df.to_csv('bitcoin_forecast.csv')
    print("Forecast data saved to 'bitcoin_forecast.csv'")

    # Save model performance metrics
    performance_df = pd.DataFrame({
        'In-Sample': pd.Series(in_sample_metrics),
        'Out-of-Sample': pd.Series(out_sample_metrics)
    })
    performance_df.to_csv('model_performance.csv')
    print("Model performance metrics saved to 'model_performance.csv'")

    print("\nProject execution completed successfully!")
    print("You can now analyze the results and interpret the forecasts.")
```
